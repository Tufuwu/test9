### Text preprocessing, TF-IDF representation and scatter visualization

```python
import texthero as hero
import pandas as pd

df = pd.read_csv(
   "https://github.com/jbesomi/texthero/raw/master/dataset/bbcsport.csv"
)

df['pca'] = (
   df['text']
   .pipe(hero.clean)
   .pipe(hero.tfidf)
   .pipe(hero.pca)
)
fig = hero.scatterplot(df, 'pca', color='topic', title="PCA BBC Sport news", return_figure=True)
fig.write_image("../github/scatterplot_bbcsport.svg")
```

### Text preprocessing, TF-IDF, K-means and visualization

```python
import texthero as hero
import pandas as pd

df = pd.read_csv(
    "https://github.com/jbesomi/texthero/raw/master/dataset/bbcsport.csv"
)

df['tfidf'] = (
    df['text']
    .pipe(hero.clean)
    .pipe(hero.tfidf)
)

df['kmeans_labels'] = (
    df['tfidf']
    .pipe(hero.kmeans, n_clusters=5)
    .astype(str)
)

df['pca'] = (
    df['tfidf']
    .pipe(hero.pca)
)

fig = hero.scatterplot(df, 'pca', color='kmeans_labels', title="K-means BBC Sport news", return_figure=True)
fig.write_image("../github/scatterplot_bbcsport_kmeans.svg")
```
